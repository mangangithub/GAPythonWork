<li>Otherwise if the time of day is before 10 p.m. --> print the quote "A day without sunshine is like, you know, night."

<li>Otherwise, for any time 10 p.m. or later --> print the quote "Burning the midnight oil!"
</ol>

```python
#problem1.py - Problem 1: Most Clocks Are Normal, But Some Are Cuckoo!

time = 21 #Military Time - 24 hours
print(f"Military Time: {time}:00\n")

if time < 9:
    print("Morning is wonderful. Its only drawback is that it comes at such an inconvenient time of day.")
elif (time >= 9 and time <= 16):
    print("Working hard or hardly working?")
elif time > 16 and time < 20:
    print("How did it get so late so soon?")
elif time >= 20 and time < 22:
    print("A day without sunshine is like, you know, night.")
else:
    print("Burning the midnight oil!")
```

# HW1_unit-2-control-flow - problem2.py

The mighty Tyrannosaurus rex is the king of dinosaurs, and he does whatever he pleases. When he's hungry, he eats. When he's angry, he fights. When he's bored, he roars. When he's tired, he sleeps! Write a conditional statement that selects one of the following actions for T-Rex based on his current mood. T-Rex's actions are ordered by precedence:<br> <br> 
If T-Rex is angry, hungry, and bored he will eat the Triceratops.<br> 
Otherwise if T-Rex is tired and hungry, he will eat the Iguanadon. <br> 
Otherwise if T-Rex is hungry and bored, he will eat the Stegasaurus. <br> 
Otherwise if T-Rex is tired, he goes to sleep.<br> 
Otherwise if T-Rex is angry and bored, he will fight with the Velociraptor.<br>
Otherwise if T-Rex is angry or bored, he roars.<br> 
Otherwise T-Rex gives a toothy smile.<br> 

```python
#problem2.py - Problem 2: I Came, I 'Saur, I Conquered

angry = True 
bored = False 
hungry = False 
tired = False

print("TRex's action based on his mood...")
if angry and hungry and bored:
    print("T-Rex is angry, hungry and bored - he eats the Triceratopos.")
elif tired and hungry:
    print("T-Rex is angry, hungry and bored - he eats the Iguanadon.")
elif hungry and bored:
    print("T-Rex is angry, hungry and bored - he eats the Stegasaurus.")
elif tired:
    print("T-Rex is tired - he sleeps")    
elif angry and bored:
    print("T-Rex is angry, hungry and bored - he eats the Velociraptor.")    
elif angry or bored:
    print("T-Rex is angry or bored - he roars.")    
else:
    print("-Rex gives a toothy smile.")        
```
